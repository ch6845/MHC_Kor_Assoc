# Imputation CookHLA included in HATK

HATK returns `*.ped` and `*.alleles`

# hped to bmarker

## load data

### Imputation output

````python
allele=pd.read_csv("/data/wschoi/ch6845_cowork/KCHIP_ch6845/CookHLA_result_HAN_China_v5/finition/KCHIP_HLA.hg18.intersection_HAN.LABELED.MERGED.MHC.HLA_IMPUTATION_OUT.alleles",sep=' ',header=None)
print(allele.shape)
allele.head()
````

````python
hped=pd.read_csv("/data/wschoi/ch6845_cowork/KCHIP_ch6845/CookHLA_result_HAN_China_v5/finition/KCHIP_HLA.hg18.intersection_HAN.LABELED.MERGED.MHC.HLA_IMPUTATION_OUT.hped",sep='\t',header=None)
print(hped.shape)
hped.head()
````

### cped example file

````python
cped_example=pd.read_csv("HATK/example/wtccc_filtered_58C_RA.hatk.300+300.imgt3320.2field.chped",sep='\t',header=None)
print(cped_example.shape)
cped_example.head()
````

## formatting (hped to chped)

````python
chped=hped.copy()
for i in range(6,chped.shape[1]):
    allele_type=cped_example[i][0].split('*')[0]

    four_field_first=(chped[i]/100).astype(int).astype(str).str.zfill(2)
    four_field_second=(chped[i]%100).astype(int).astype(str).str.zfill(2)

    chped_format=allele_type+'*'+four_field_first+':'+four_field_second
    chped[i]=chped_format
````

````python
chped.head()
````

## saving chped

````python
chped.to_csv('data/genotype/hped/KCHIP.chped',sep='\t',header=None,index=None)
````
